```python
import pandas as pd
import numpy as np
df = pd.read_csv("exoplanet_data.csv")
```

# Read the CSV and Perform Basic Data Cleaning

```python
# Exploratory data analysis
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
df
```

```python
# Inspect rows and columns with missing data and decide what to do with them

# Drop rows/columns with missing data or you can replace missing data
X = df.drop("koi_disposition",axis=1)
```

# Select your features (columns)

```python
# Set your y to be poi-disposition and your X to be all other columns
y=df['koi_disposition']
print(X.shape)
print(y.shape)
```

```python
# Drop *unimportant, irrelevant, or duplicate columns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
```

# Create a Train Test Split


```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

# Pre-processing

```python
# Scale your data using appropriate scaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train various  models



## Model 1

```python
# Deep Learning with 3 hidden layers each having 10 inputs

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create a Neural Network model here
model=Sequential()
model.add(Dense(units=10, activation='relu', input_dim=40))
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=3, activation='softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=500,
    shuffle=True,
    verbose=2
)
```

```python
dl_model_loss, dl_model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Normal Neural Network - Loss: {dl_model_loss}, Accuracy: {dl_model_accuracy}")
```

```python
encode=model.predict_classes(X_test[:5])
dl_prediction_labels = label_encoder.inverse_transform(encode)
print(dl_prediction_labels)
```
